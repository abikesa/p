The idea that bureaucrats "destroy grammar by transforming verbs into nouns" is a somewhat exaggerated and generalized statement, but it touches on a linguistic phenomenon often associated with bureaucratese or legalese. This phenomenon is called "nominalization."

Nominalization refers to the process of turning verbs (or adjectives) into nouns. For example, the verb "decide" can be turned into the noun "decision." While nominalization is a standard feature of the English language and is used in various contexts, it is often criticized in bureaucratic writing for several reasons:

1. **Complexity**: Bureaucratic documents frequently use nominalizations, which can make sentences longer and harder to understand. For instance, "make a decision" instead of "decide" or "provide an explanation" instead of "explain."

2. **Passivity**: Nominalization can make sentences more passive, distancing the action from the doer. This can make sentences seem more impersonal or vague about who is responsible for an action.

3. **Ambiguity**: Over-reliance on nominalizations can lead to ambiguous statements. The reader might have difficulty understanding who did what and why.

However, it's essential to understand that nominalization in itself is not inherently bad. It can be useful for abstraction, generalization, or emphasizing particular aspects of an action. The problem arises when it's overused or used unnecessarily, leading to convoluted and unclear language.

In summary, while bureaucrats (and other professionals, like lawyers) might often use nominalizations in their writing, it's not accurate to say they "destroy grammar." Instead, the overuse of nominalization can make their language less direct and harder to understand. Proper training in clear communication can help mitigate these issues.
